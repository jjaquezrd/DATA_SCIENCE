```python
import pandas as pd
url_archivo='https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data'
df = pd.read_csv(url_archivo, header = None)
df
```

```python
df.info()
```

Renombrar columnas
Renombrar columna 0 a “Flow_ID”
Renombrar columna 1 a “Date”
Renombrar columna 2 a “Time”
Renombrar columna 3 a “Count”

```python
df = df.rename(columns={0:'Flow_ID',1:'Date',2:'Time',3:'Count'})
df.head()
```

Seleccionar todos los datos para la fecha 24 de julio de 2005 con un flow id=7.

```python
fecha_filtrada = df['Date']=='07/24/05'
fecha_filtrada
```

```python
flow_id_filtrado = df['Flow_ID']==7
flow_id_filtrado
```

```python
df_filtrado = df.loc[fecha_filtrada & flow_id_filtrado,:]
df_filtrado.head()
```

Con dos datos seleccionados en el paso 1, ordenen el resultado en Count en orden descendiente y saquen las 10 primeras filas. Asignen la salida a una variable llamada top 10.

```python
top10=df_filtrado.sort_values('Count', ascending=False)
top10=top10.head(10)
top10
```

Para las 10 filas que aparecen arriba, utilice la función Apply de Pandas para restar el valor más bajo de los 10 de todos ellos y luego obtengan el valor promedio de los recuentos resultantes

```python
def rest_min(x, frame,columna):
  restando =frame[str(columna)].min()
  return x-int(restando)


```

```python
top10['Count']=top10['Count'].apply(rest_min, args=[top10,'Count'])
```

```python
top10.mean()
```

Seleccionar los datos en el mes de agosto 2005 con un flow id=7
Agrupar los datos en función de la fecha y obtener el recuento máximo por fecha

```python
df.info()
```

```python
##df['Date'] = pd.to_datetime(df['Date'],format='%Y-%m-%d')
agosto2005= df.loc[(df['Date'] >='08/01/2005')
                   &(df['Date'] <='08/31/2005')
                   & (df['Flow_ID'] ==7)]
```

```python
agosto2005
```

```python
agosto2005.groupby('Date')[['Count']].max()
```
